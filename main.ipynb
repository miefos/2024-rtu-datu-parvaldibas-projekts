```python
import pandas as pd
file_path = 'data/data_prepared.xlsx'

df = pd.read_excel(file_path, header=0)
df
```

```python
def standardize_number(value):
    return float(str(value).replace(' ', '').replace(',', '.'))

for col in df.columns[1:]:
    df[col] = df[col].apply(standardize_number)

df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import os

# Ensure the charts folder exists
charts_folder = 'charts/original'
if not os.path.exists(charts_folder):
    os.makedirs(charts_folder)

# Iterate through all cols except first one
for col in df.columns[1:]:
    fig, ax = plt.subplots(1, 2, figsize=(14,7))  # 1 row, 2 columns

    # Bar chart on the left (axis 0)
    df.plot.bar(x='year', y=col, ax=ax[0])
    ax[0].set_title(f"{col} Bar Chart")

    # Line chart on the right (axis 1)
    df.plot(x='year', y=col, ax=ax[1])
    ax[1].set_title(f"{col} Line Chart")

    # Adjust spacing and save
    plt.tight_layout()
    location = os.path.join(charts_folder, f"{col}_combined.png")
    plt.savefig(location)
    plt.close()

    print(f"saved: {location}")
```

```python
from sklearn.preprocessing import StandardScaler
# Goal: Transform data to have a mean of 0 and a standard deviation of 1.
# This is often done to make data unitless and comparable across different scales.
# Z-Score standardization

# Scale everything except year column
df_scaled = df.copy()
cols_to_scale = [col for col in df.columns if col != 'year']
scaler = StandardScaler()
df_scaled[cols_to_scale] = scaler.fit_transform(df_scaled[cols_to_scale])
df_scaled
```

```python
import seaborn as sns
# Pearson correlation coefficients:
# The Pearson correlation coefficient (r) is the most common way of measuring a linear correlation.
# It is a number between –1 and 1 that measures the strength and direction of the relationship between two variables.
# corr() by default uses Pearson.
correlations = df_scaled[df_scaled.columns[1:]].corr()

plt.figure(figsize=(16, 8))
corr_matrix = correlations
sns.heatmap(corr_matrix, annot=True, cmap='RdBu', vmin=-1, vmax=1, fmt='.0%')
plt.title('Correlation Coefficients')
```

1. Elektrības cenai ir liela korelācija ar elektroauto reģistrācijām. Tomēr, tas visticamāk nav cēlonis tam, ka tiek reģistrētas vairāk elektroauto. Drīzāk būtu jābūt tieši otrādi. Tā vienkārši sagadījies, ka liels elektroauto reģistrēto skaits sakrita ar strauji augošu elektrības cenu.

```python
# 1. Multiple Linear Regression without train / test
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Prepare X and y
X = df_scaled.drop(['el_auto_registr', 'year'], axis=1)
y = df_scaled['el_auto_registr']

# Train the model
model = LinearRegression()
model.fit(X, y)

# Make prediction
predictions = model.predict(X)

# MSE and R^2
mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)

print('Metrics:')
print('Mean Squared Error:', mse)
print('R-squared:', r2)
print('')
print('Results comparison in table:')

# Set predictions, inversed scaler
df_scaled_predictions = df_scaled.copy()
df_scaled_predictions['el_auto_registr'] = predictions

df_scaled_predictions[cols_to_scale] = scaler.inverse_transform(df_scaled_predictions[cols_to_scale])

# Compare results in a table
comparison_mlr_results = df.copy()
comparison_mlr_results.insert(2, 'predictions', df_scaled_predictions['el_auto_registr'].to_list())
comparison_mlr_results.insert(3, 'diff', (df_scaled_predictions['el_auto_registr'].to_list() - comparison_mlr_results['el_auto_registr']).round(2))
comparison_mlr_results.insert(4, 'diff %', ((df_scaled_predictions['el_auto_registr'].to_list() - comparison_mlr_results['el_auto_registr']) / comparison_mlr_results['el_auto_registr']).round(2))

print(comparison_mlr_results[['year', 'el_auto_registr', 'predictions', 'diff', 'diff %']])

# Plot results
real_values = df['el_auto_registr']
predicted_values = df_scaled_predictions['el_auto_registr']

real_colors = ['blue' for _ in real_values]
predicted_colors = ['red' for _ in predicted_values]

plt.figure(figsize=(10, 6))
plt.scatter(df['year'], real_values, label='Real Values', color=real_colors)
plt.scatter(df['year'], predicted_values, label='Predicted Values', color=predicted_colors)

plt.xlabel('Year')
plt.ylabel('el_auto_registr')
plt.title('Real vs. Predicted el_auto_registr')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# 1. Multiple Linear Regression with train / test
from sklearn.model_selection import train_test_split

# Prepare X and y
X = df_scaled.drop(['el_auto_registr'], axis=1)
y = df_scaled['el_auto_registr']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=0)

train_years = X_train['year']
test_years = X_test['year']

X_train = X_train.drop(['year'], axis=1)
X_test = X_test.drop(['year'], axis=1)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make prediction
predictions = model.predict(X_test)

# Set predictions, inversed scaler
df_scaled_predictions = df_scaled[df_scaled['year'].isin(test_years)].copy()
df_scaled_predictions['el_auto_registr'] = predictions

df_scaled_predictions[cols_to_scale] = scaler.inverse_transform(df_scaled_predictions[cols_to_scale])

# Compare results in a table
comparison_mlr_results = df[df['year'].isin(test_years)].copy()
comparison_mlr_results.insert(2, 'predictions', df_scaled_predictions['el_auto_registr'].to_list())
comparison_mlr_results.insert(3, 'diff', (df_scaled_predictions['el_auto_registr'].to_list() - comparison_mlr_results['el_auto_registr']).round(2))
comparison_mlr_results.insert(4, 'diff %', ((df_scaled_predictions['el_auto_registr'].to_list() - comparison_mlr_results['el_auto_registr']) / comparison_mlr_results['el_auto_registr'] * 100).round(2))

# MSE and R^2
mse = mean_squared_error(comparison_mlr_results['el_auto_registr'], comparison_mlr_results['predictions'])
r2 = r2_score(comparison_mlr_results['el_auto_registr'], comparison_mlr_results['predictions'])

print('Metrics:')
print('Mean Squared Error:', mse)
print('R-squared:', r2)
print('')
print('Results comparison in table:')

print(comparison_mlr_results[['year', 'el_auto_registr', 'predictions', 'diff', 'diff %']])

# Plot results
real_values = df[df['year'].isin(test_years)]['el_auto_registr']
predicted_values = df_scaled_predictions['el_auto_registr']

real_colors = ['blue' for _ in real_values]
predicted_colors = ['red' for _ in predicted_values]

plt.figure(figsize=(10, 6))
plt.scatter(test_years, real_values, label='Real Values', color=real_colors)
plt.scatter(test_years, predicted_values, label='Predicted Values', color=predicted_colors)

plt.xlabel('Year')
plt.ylabel('el_auto_registr')
plt.title('Real vs. Predicted el_auto_registr')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# 2. XGBoost with train / test
import xgboost as xgb

# Prepare X and y
X = df_scaled.drop(['el_auto_registr'], axis=1)
y = df_scaled['el_auto_registr']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=0)

train_years = X_train['year']
test_years = X_test['year']

X_train = X_train.drop(['year'], axis=1)
X_test = X_test.drop(['year'], axis=1)

# Train the model
model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='rmse')
model.fit(X_train, y_train)

# Make prediction
predictions = model.predict(X_test)

# Set predictions, inversed scaler
df_scaled_predictions = df_scaled[df_scaled['year'].isin(test_years)].copy()
df_scaled_predictions['el_auto_registr'] = predictions

df_scaled_predictions[cols_to_scale] = scaler.inverse_transform(df_scaled_predictions[cols_to_scale])

# Compare results in a table
comparison_mlr_results = df[df['year'].isin(test_years)].copy()
comparison_mlr_results.insert(2, 'predictions', df_scaled_predictions['el_auto_registr'].to_list())
comparison_mlr_results.insert(3, 'diff', (df_scaled_predictions['el_auto_registr'].to_list() - comparison_mlr_results['el_auto_registr']).round(2))
comparison_mlr_results.insert(4, 'diff %', ((df_scaled_predictions['el_auto_registr'].to_list() - comparison_mlr_results['el_auto_registr']) / comparison_mlr_results['el_auto_registr'] * 100).round(2))

# MSE and R^2
mse = mean_squared_error(comparison_mlr_results['el_auto_registr'], comparison_mlr_results['predictions'])
r2 = r2_score(comparison_mlr_results['el_auto_registr'], comparison_mlr_results['predictions'])

print('Metrics:')
print('Mean Squared Error:', mse)
print('R-squared:', r2)
print('')
print('Results comparison in table:')

print(comparison_mlr_results[['year', 'el_auto_registr', 'predictions', 'diff', 'diff %']])

# Plot results
real_values = df[df['year'].isin(test_years)]['el_auto_registr']
predicted_values = df_scaled_predictions['el_auto_registr']

real_colors = ['blue' for _ in real_values]
predicted_colors = ['red' for _ in predicted_values]

plt.figure(figsize=(10, 6))
plt.scatter(test_years, real_values, label='Real Values', color=real_colors)
plt.scatter(test_years, predicted_values, label='Predicted Values', color=predicted_colors)

plt.xlabel('Year')
plt.ylabel('el_auto_registr')
plt.title('Real vs. Predicted el_auto_registr')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
